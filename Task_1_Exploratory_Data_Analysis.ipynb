# **Weak Areas for More Profit:**

Analyze the 'Profit' column. Calculate the total profit, average profit, and profit distribution across different categories, sub-categories, and regions.

```python
total_profit = df["Profit"].sum()
avg_profit = df["Profit"].mean()

print("Total Profit : ",total_profit," & Average Profit : " , avg_profit)
```

```python
profit_by_category = df.groupby("Category")["Profit"].sum().sort_values(ascending = False)
profit_by_category
```

```python
profit_by_subcategory = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending = False)
print(profit_by_subcategory,"\n")

plt.figure(figsize=(18,10))
sns.barplot(x = profit_by_subcategory.index , y = profit_by_subcategory.values )
plt.ylabel("Profit")
plt.title("Profit Distribution by  Subcategory ")
plt.show()
```

```python
profit_by_region = df.groupby("Region")["Profit"].sum().sort_values(ascending = False)
profit_by_region
```

```python
category_region_profit = df.groupby(["Category" , "Region"])['Profit'].sum().reset_index()
category_region_profit
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x="Category", y="Profit", hue="Region", data=category_region_profit)
plt.title("Profit by Category and Region")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation=0)
plt.legend(title="Region")
plt.show()
```

# **Therefore Central region has low profit & Technology category can increase profit.**
